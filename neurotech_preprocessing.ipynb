```python
import matplotlib
import pathlib
import pandas as pd
import mne
import numpy as np
```

```python
path = '/Users/aryamanbhatia/mne_data/MNE-sample-data/neurotech-data/OpenBCISession_2023-03-03_15-55-03/OpenBCI-RAW-2023-03-03_15-56-44.txt'
raw = pd.read_csv(path, delimiter=',', skiprows=4)
raw
```

```python
headers = list(raw.columns.values)
ch_names = headers[1:9]

ch_names
```

```python
#df = pd.read_csv(path, delimiter = ',', skiprows=4) 
data = raw.to_numpy()
headers = list(raw.columns.values)

n_channels = 8
sampling_freq = 250
ch_names = headers[1:9] #or do it manually with header

info = mne.create_info(ch_names = ch_names, sfreq=sampling_freq, ch_types = ['eeg'] * 8)
info

len(info['ch_names'])

data_t = data[:,1:9].transpose()
raw = mne.io.RawArray(data_t, info)
#raw.plot()
#mne.filter.create_filter(raw, 200, 1, 15)
l_freq = 8
h_freq = 32
raw_filt = raw.copy().filter(l_freq = l_freq, h_freq = h_freq)

data_t.shape

evoked_array = mne.EvokedArray(data_t, info, tmin=-0.1,
                               nave=9, kind='standard_error')
print(evoked_array)
evoked_array.plot()
```
